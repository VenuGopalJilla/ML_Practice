```python
p = {}
for i in train_df.columns:
    contingency_table=pd.crosstab(train_df[i],train_df['revenue'])
    Observed_Values = contingency_table.values 

    import scipy.stats
    b=scipy.stats.chi2_contingency(contingency_table)     # ((sumof row)*(sum of column))/(total)
    Expected_Values = b[3]

    no_of_rows=len(contingency_table.iloc[0:2,0])
    no_of_columns=len(contingency_table.iloc[0,0:2])
    df=(no_of_rows-1)*(no_of_columns-1)
    
    from scipy.stats import chi2
    chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
    chi_square_statistic=chi_square[0]+chi_square[1]

    p_value=1-chi2.cdf(x=chi_square_statistic,df=df)
    print('p-value of '+i+" : ",p_value)
    p[i] = p_value

# print(p)
max_val = max(p, key = p.get)
print("max value : " , max_val + " , " ,  p[max_val])
```
